```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('tips.csv')
df.head()
```

```python
sns.catplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,jitter=False)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='swarm')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,hue='sex',kind='swarm')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,hue='time',kind='swarm')
plt.show()
```
