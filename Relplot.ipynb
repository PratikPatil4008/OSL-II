```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('tips.csv')
df.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker',style='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker',style='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='size')
plt.show()
```
