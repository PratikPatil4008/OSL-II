```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('titanic.csv')
df.head()
```

```python
sns.catplot(x='sex',y='survived',data=df,hue='class',kind='point')
plt.show()
```

```python
sns.catplot(x='class',y='survived',data=df,hue='sex',kind='point',palette={'male':'g','female':'m'},markers=['^','o'],linestyle=['-','--'])
plt.show()
```
