```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('tips.csv')
df.head()
```

```python
sns.catplot(x='total_bill',y='day',hue='time',kind='violin',data=df)
plt.show()
```

```python
sns.catplot(x='total_bill',y='day',hue='time',kind='violin',bw=.15,cut=0,data=df)
plt.show()
```
